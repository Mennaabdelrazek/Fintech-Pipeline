Since Branch is Unique within a City, we can use cities to impute the Branch

```python
# Impute missing branch with city
# Get mapping from city_branch
city_branch_dict = {city: branch for city, branch in city_branch.index}
city_branch_dict
```

```python
# Impute missing branch
sales_df['branch'] = sales_df['branch'].fillna(sales_df['city'].map(city_branch_dict))
sales_df.isnull().sum()
```

```python
# Impute tax 5% by calculating it
# Calculate tax 5%
sales_df.head()
```

```python
def calculate_tax(price, quantity):
    return (price * quantity) * 0.05

# Apply calculate_tax function on null values
sales_df['tax_5%'] = sales_df['tax_5%'].fillna(sales_df.apply(lambda x: calculate_tax(x['unit_price'], x['quantity']), axis=1))
sales_df.head()
```

```python
sales_df.isnull().sum()
```

```python
# Now pack it into functions
def impute_missing(df):
    # Impute missing branch
    city_branch = df.groupby('city')['branch'].value_counts()
    city_branch_dict = {city: branch for city, branch in city_branch.index}
    df['branch'] = df['branch'].fillna(df['city'].map(city_branch_dict))

    # Impute missing tax 5%
    df['tax_5%'] = df['tax_5%'].fillna(df.apply(lambda x: calculate_tax(x['unit_price'], x['quantity']), axis=1))

    return df
```

```python
# Encoding using Label Encoding: Branch, City, Gender, Product Line

from sklearn.preprocessing import LabelEncoder

label_encoders = {}
columns_to_encode = ['branch', 'city', 'gender', 'product_line']

for column in columns_to_encode:
    le = LabelEncoder()
    sales_df[column] = le.fit_transform(sales_df[column])
    label_encoders[column] = le
    # Save the mapping
    mapping_df = pd.DataFrame({
        'Column': [column] * len(le.classes_),
        'Old Value': le.classes_,
        'New Value': range(len(le.classes_))
    })
    lookup = pd.concat([lookup, mapping_df], ignore_index=True)

sales_df.head()
```

```python
# Encoding using One Hot Encoding: Customer Type (is_member), Payment

sales_df['is_member'] = sales_df['customer_type'].apply(lambda x: 1 if x == 'Member' else 0)
sales_df.drop(columns=['customer_type'], inplace=True)

# One Hot Encoding 0,1
sales_df = pd.get_dummies(sales_df, columns=['payment'], drop_first=True)
sales_df.columns = sales_df.columns.str.lower().str.replace(' ', '_')
sales_df.replace({False: 0, True: 1}, inplace=True)
```

```python
def encoding(df):
    global lookup
    lookup = pd.DataFrame(columns=['Column', 'Old Value', 'New Value'])

    # Label Encoding
    label_encoders = {}
    columns_to_encode = ['branch', 'city', 'gender', 'product_line']

    for column in columns_to_encode:
        le = LabelEncoder()
        df[column] = le.fit_transform(df[column])
        label_encoders[column] = le
        # Save the mapping
        mapping_df = pd.DataFrame({
            'Column': [column] * len(le.classes_),
            'Old Value': le.classes_,
            'New Value': range(len(le.classes_))
        })
        lookup = pd.concat([lookup, mapping_df], ignore_index=True)

    # One Hot Encoding
    df['is_member'] = df['customer_type'].apply(lambda x: 1 if x == 'Member' else 0)
    df.drop(columns=['customer_type'], inplace=True)

    # One Hot Encoding 0,1
    df = pd.get_dummies(df, columns=['payment'], drop_first=True)
    df.columns = df.columns.str.lower().str.replace(' ', '_')
    df.replace({False: 0, True: 1}, inplace=True)

    return df
```

```python
# Save the lookup table and the label encoders
lookup.to_csv('lookup.csv', index=False)
sales_df.to_csv('sales_clean.csv', index=False)
```
